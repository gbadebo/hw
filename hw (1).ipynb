```python
import pandas as pd
import json
```

```python
 csv_df = pd.read_csv("csv_file.csv")
 txt_df = pd.read_csv("txt_file.txt", sep=",")
 json_data = open("json_file.json")
 json_df = json.load(json_data)
json_df_normalized = pd.json_normalize(json_df)
json_df_normalized
```

```python
csv_txt_merged = pd.concat([csv_df, txt_df])
```

```python
csv_txt_json_merged = pd.concat([csv_txt_merged, json_df_normalized])
csv_txt_json_merged.shape
```

```python
!pip install autogluon
```

```python
csv_txt_json_merged.rename(columns = {'author.properties.friends': 'friends',
                        'author.properties.status_count': 'status_count',
                        'author.properties.verified': 'verified', 
                        'content.body': 'text', 
                        'location.country': 'country',
                        'properties.platform': 'platform', 
                        'properties.sentiment': 'sentiment', 
                        'location.latitude': 'latitude',
                        'location.longitude': 'longitude'},
                        inplace=True)


df = csv_txt_json_merged[['text', 'sentiment']]
df.head()
```

```python
missingdata = df[df.isnull().any(axis=1)]
missingdata
df = df.dropna(axis=0)
```

```python
import autogluon as ag
```

```python
df["sentiment"] = df["sentiment"].astype('int64')
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
!pip install autogluon.text
```

```python



train_data, test_data = train_test_split(
    df, test_size=0.3, random_state=42
)
from autogluon.text import TextPredictor

predictor = TextPredictor(label='sentiment', eval_metric='acc', path='artifact1s/model')
predictor.fit(train_data, time_limit=60)
```

```python
test_score = predictor.evaluate(test_data)
print(test_score)
print('Accuracy = {:.2f}%'.format(test_score['acc'] * 100))
```
